```python
import xml.etree.ElementTree as ET
import copy
```

```python
tree = ET.parse('antibody_sentences_final_subset.xml')
root = tree.getroot()
```

```python
interesting_frames = ['Assistance','Supply','Ingredients','Locale_by_use','Supporting'] 
```

```python
sentences = []
for child in root:
    for child1 in child:
        for child2 in child1:
            for child3 in child2:
                for child4 in child3:
                    for child5 in child4:
                        sentences.append(child5)
```

```python
f = open('semafor_results.txt','w')
f2 = open('donors.tsv','w')
f2.write('Sentence number \t Donor \t Sentence \n')
f3 = open('antibodies.tsv','w')
f3.write('Sentence number \t Antibody \t Sentence \n')
for ind,sentence in enumerate(sentences):
    for child in sentence:
        if child.text.strip() != '':
            sentence_text = child.text
            print sentence_text
            f.write('=============================================================================================\n')
            f.write(sentence_text+'\n')
        for child1 in child: # Child1 contains the name of the frame
            if child1.attrib['frameName'] not in interesting_frames:
                continue
            supplier = ''
            theme = ''
            frame_name = child1.attrib['frameName']
            print '======================================================'
            f.write('======================================================\n')
            print 'Found a Frame : ' + frame_name
            f.write('Found a Frame : ' + frame_name + '\n')
            print '======================================================'
            f.write('======================================================\n')
            for child2 in child1: # Child 2 is nothing. 
                for child3 in child2: #Child 3 contains whether a word in the frame is a FE or a target.
                    word_type = child3.attrib['name']
                    print 'Found a : ' + word_type
                    f.write('Found a : ' + word_type +'\n')
                    for child4 in child3:
                        for child5 in child4:
                            start = int(child5.attrib['start'])
                            end = int(child5.attrib['end'])
                            print 'Lexicon : ' + sentence_text[start:end+1] + ' Role : ' + child5.attrib['name']
                            f.write('Lexicon : ' + sentence_text[start:end+1] + ' Role : ' + child5.attrib['name'] + '\n')
                            if child5.attrib['name'] == 'Supplier' and frame_name == 'Supply':
                                supplier = sentence_text[start:end+1]
                                f2.write(str(ind) + '\t' + sentence_text[start:end+1] + '\t' + sentence_text.replace('\t','') + '\n')
                            if child5.attrib['name'] == 'Theme' and frame_name == 'Supply':
                                theme = sentence_text[start:end+1]
                                f3.write(str(ind) + '\t' + sentence_text[start:end+1] + '\t' + sentence_text.replace('\t','') + '\n')
```
